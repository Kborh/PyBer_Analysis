```python
# import dependency
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot
plt.plot(x_axis , y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot
plt.bar(x_axis, y_axis, color="green", label='Buston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt() using object-oriented intrface
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label='Chicago')
#create plot lebal
ax.set_xlabel("Faire($)")
ax.set_ylabel("Date")
#Create limit between 0 and 45
ax.set_xlim(0,45)
#Create a title
ax.set_title("PyBer Fare by Month")
ax.invert_yaxis()
ax.legend()
```

```python
#Create a scatter plot with MATPLOT
# Create the plot
plt.scatter(x_axis, y_axis, s=y_axis, color="green", label='Buston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#Adjust the size of the data on the y_axis
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
#create a scattter plot using the MATLAB method
plt.figure()
plt.scatter(x_axis, y_axis, s=y_axis_larger, c="cyan", label="Boston")
#create title
plt.title('PyBer Fare by Month')
#create lable for y and x
plt.ylabel('Fate($)')
plt.xlabel('Date')
#Add limitation to x_axis between 0 and 50
plt.ylim(0,50)
#Invert the y_axis
#plt.gca().invert_yaxis()
#add legand
plt.legend()
plt.show()
```

```python
#Using python list comprehision to refactor the code
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
#Create scatter plot using the object-oriented interface
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis],edgecolors= "black",alpha=0.3, linewidth=2, label='Boston')
#Create a title
ax.set_title("PyBer Fare by Month")
#Create lable for x and y
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
#Add legand
ax.legend()
```
